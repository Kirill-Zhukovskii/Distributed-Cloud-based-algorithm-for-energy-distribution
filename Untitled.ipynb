```python
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cp
from scipy.linalg import block_diag
import copy
```

```python
# Set seed for reproducibility
np.random.seed(42)

# === 1. Time Horizon ===
H = 24
N=10
hours = np.arange(H)
base_price = 0.15
amplitude = 0.10
noise = np.random.normal(0, 0.01, size=H)

# Symmetric price
prices = base_price + amplitude * (
    np.sin((hours - 8) / 24 * 2 * np.pi) + np.sin((hours - 18) / 24 * 2 * np.pi)
) + noise
prices = np.clip(prices, 0.05, None)

# Add asymmetric buy/sell prices
margin = 0.05
c = prices + margin
p = prices - margin
p = np.clip(p, 0.01, None)  # Prevent negative prices


# === 3. Generate Renewable Energy ===
pv_gen = np.maximum(0, 5 * np.sin((hours - 6) * np.pi / 12))  # PV peaks at noon
wt_gen = 2 + np.random.rand(H) * 2  # Wind is random but consistent
q = (pv_gen + wt_gen)


# Plotting
fig, axs = plt.subplots(2, 1, figsize=(10, 12.5))  # 1 row, 2 columns

# First subplot: Buy and Sell Prices
axs[0].plot(hours, c, label='Buy Price (€/kWh)')
axs[0].plot(hours, p, label='Sell Price (€/kWh)')
axs[0].plot(hours, prices, label='Base Market Price (€/kWh)', linestyle='--', alpha=0.7)
axs[0].set_title("Simulated 24-Hour Electricity Buy and Sell Prices")
axs[0].set_xlabel("Hour")
axs[0].set_ylabel("Price (€/kWh)")
axs[0].grid(axis='y')
axs[0].set_xticks(range(0, 24))
axs[0].legend()

# Second subplot: Renewable energy
axs[1].plot(np.arange(H), q)
axs[1].set_title("Renewable Energy")
axs[1].set_xlabel("Hour")
axs[1].set_ylabel("Energy (Wh)")
axs[1].grid(axis='y')
axs[1].set_xticks(range(0, 24))

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import cvxpy as cp
import copy

# Problem parameters
N = 5  # Number of agents
H = 10  # Horizon length
rho = 100
epsilon = 1e-2
c1 = 5

# Random seeds for reproducibility
np.random.seed(42)

# Constants and bounds
L_target = np.random.uniform(70, 90, size=N)
L_init = np.random.uniform(20, 60, size=N)

L_min = np.ones((N, H)) * 20
L_max = np.ones((N, H)) * 100

d_min = np.zeros((N, H))
d_max = np.ones((N, H)) * 10

f_min = np.zeros((N, H))
f_max = np.ones((N, H)) * 10

l_min = np.zeros((N, H))
l_max = np.ones((N, H)) * 10

m_min = np.zeros((N, H))
m_max = np.ones((N, H)) * 10

# Collect data for plotting
d_history = np.array([np.zeros(H) for n in range(N)])
f_history = np.array([np.zeros(H) for n in range(N)])
l_history = np.array([np.zeros(H) for n in range(N)])
m_history = np.array([np.zeros(H) for n in range(N)])
L_history = np.array([np.zeros(H) for n in range(N)])

# Initial values
x = np.array([np.zeros(5 * H) for _ in range(N)])  # d, f, l, m, L
x_prev = copy.deepcopy(x)

z = np.zeros(H)  # Shared consensus variable
lambdas = [np.zeros(H) for _ in range(N)]  # Per-agent duals

# ADMM loop
iteration = 0
while True:
    iteration += 1
    r_n_list = []

    for n in range(N):
        xi = cp.Variable(5 * H)
        d = xi[:H]
        f = xi[H:2 * H]
        l = xi[2 * H:3 * H]
        m = xi[3 * H:4 * H]
        L = xi[4 * H:]

        # Residual expression
        r_n_expr = -d + f + l - m

        # Cost components
        net_cost = c1 * cp.sum(c @ d / q) - cp.sum(p @ f / q)
        charge_penalty = cp.square(L[-1] - L_target[n])
        admm_penalty = cp.sum_squares(r_n_expr - z + lambdas[n])

        objective = cp.Minimize(net_cost + charge_penalty + (rho / 2) * admm_penalty)

        # Constraints
        constraints = [
            L >= L_min[n],
            L <= L_max[n],
            d >= d_min[n], d <= d_max[n],
            f >= f_min[n], f <= f_max[n],
            l >= l_min[n], l <= l_max[n],
            m >= m_min[n], m <= m_max[n],
            L[0] == L_init[n] + l[0] - m[0]
        ]
        for h in range(1, H):
            constraints += [L[h] == L[h - 1] + l[h] - m[h]]

        # Solve subproblem
        prob = cp.Problem(objective, constraints)
        prob.solve(solver=cp.CVXOPT, warm_start=True)

        x[n] = xi.value
        r_n_list.append(r_n_expr.value)

                # Collect data for plotting
        d_history[n] = copy.deepcopy(d.value)
        f_history[n] = copy.deepcopy(f.value)
        l_history[n] = copy.deepcopy(l.value)
        m_history[n] = copy.deepcopy(m.value)
        L_history[n] = copy.deepcopy(L.value)

    r_n_array = np.array(r_n_list)
    z_prev = z.copy()
    z = np.mean([r_n + lambdas[n] for n, r_n in enumerate(r_n_array)], axis=0)

    # Dual updates
    for n in range(N):
        lambdas[n] += r_n_array[n] - z

    # Convergence check
    primal_res = np.linalg.norm(np.sum(r_n_array, axis=0) - q)
    dual_res = np.linalg.norm(z - z_prev)
    x_diff = np.linalg.norm(x - x_prev)

    print(f"Iter {iteration} | x_diff: {x_diff:.4e} | Primal: {primal_res:.4e} | Dual: {dual_res:.4e}")

    if x_diff < epsilon and primal_res < epsilon and dual_res < epsilon:
        break

    x_prev = copy.deepcopy(x)

# Final objective value (sum of individual costs)
final_cost = 0
for n in range(N):
    d = x[n][:H]
    f = x[n][H:2 * H]
    L = x[n][4 * H:]
    net_cost = c1 * np.sum(c * d / q) - np.sum(p * f / q)
    charge_penalty = (L[-1] - L_target[n]) ** 2
    final_cost += net_cost + charge_penalty

final_cost
```

```python
# Plotting
fig, axes = plt.subplots(5, 1, figsize=(10, 5*2.5), sharex=True)

for n in range(N):
    #axes[0].plot(np.full(H,L_target[n]))
    axes[0].plot(L_history[n], label =f'agent {n}')
    axes[0].grid(axis='y')
    axes[0].legend()
    axes[0].set_ylabel("Load, Wh*h")

for n in range(N):
    axes[1].plot(d_history[n])
    axes[1].grid(axis='y')
    axes[1].set_ylabel("Draw from external grid, Wh")

for n in range(N):
    axes[2].plot(f_history[n])
    axes[2].grid(axis='y')
    axes[2].set_ylabel("Feed to external grid, Wh")
    
for n in range(N):
    axes[3].plot(m_history[n])
    axes[3].grid(axis='y')
    axes[3].set_ylabel("Discharging to local DC-grid, Wh")
    
for n in range(N):
    axes[4].plot(l_history[n])
    axes[4].grid(axis='y')
    axes[4].set_ylabel("Charging from local DC-grid, Wh")
    

axes[-1].set_xlabel('Time Horizon, hour')
plt.tight_layout()
plt.show()
```
