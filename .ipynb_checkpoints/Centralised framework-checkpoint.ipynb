```python
import matplotlib.pyplot as plt
import numpy as np
import cvxpy as cp
```

```python
# Set seed for reproducibility
np.random.seed(42)

# === 1. Time Horizon ===
H = 24
horizon = 6
N=10
hours = np.arange(H)
base_price = 0.15
amplitude = 0.10
noise = np.random.normal(0, 0.01, size=H)

# Symmetric price
prices = base_price + amplitude * (
    np.sin((hours - 8) / 24 * 2 * np.pi) + np.sin((hours - 18) / 24 * 2 * np.pi)
) + noise
prices = np.clip(prices, 0.05, None)

# Add asymmetric buy/sell prices
margin = 0.05
c = prices + margin
p = prices - margin
p = np.clip(p, 0.01, None)  # Prevent negative prices

# Plotting
plt.figure(figsize=(10, 4))
plt.plot(hours, c, label='Buy Price (/kWh)', marker='o')
plt.plot(hours, p, label='Sell Price (/kWh)', marker='s')
plt.plot(hours, prices, label='Base Market Price (/kWh)', linestyle='--', alpha=0.7)
plt.title("Simulated 24-Hour Electricity Buy and Sell Prices")
plt.xlabel("Hour")
plt.ylabel("Price (/kWh)")
plt.grid(True)
plt.xticks(range(0, 24))
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# === 3. Generate Renewable Energy ===
pv_gen = np.maximum(0, 5 * np.sin((hours - 6) * np.pi / 12))  # PV peaks at noon
wt_gen = 2 + np.random.rand(H) * 2  # Wind is random but consistent
q = pv_gen + wt_gen

plt.figure(figsize=(10, 4))
plt.plot(np.arange(H), q, marker='o')
plt.title("Renewable energy")
plt.xlabel("Hour")
plt.ylabel("Price (Wh)")
plt.grid(True)
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()
```

```python
# Parameters
N = 3        # Number of agents


L_target = np.ones(N) * 10
L_init = np.ones(N) * 5

L_min = np.zeros((N, H))
L_max = np.ones((N, H)) * 20

d_min = np.zeros((N, H))
d_max = np.ones((N, H)) * 10

f_min = np.zeros((N, H))
f_max = np.ones((N, H)) * 10

# Decision variables for each agent
x = [cp.Variable(3*H) for _ in range(N)]

constraints = []
objective = 0
for n in range(N):
    d = x[n][:H]
    f = x[n][H:2*H]
    L = x[n][2*H:]

    constraints += [
        L >= L_min[n],
        L <= L_max[n],
        d >= d_min[n],
        d <= d_max[n],
        f >= f_min[n],
        f <= f_max[n]
    ]

    constraints += [
        L[0] == L_init[n] + d[0] - f[0]
    ]
    for h in range(1, H):
        constraints += [
            L[h] == L[h-1] + d[h] - f[h]
        ]

    cost_term = cp.sum(c @ d / q) - cp.sum(p @ f / q)
    penalty_term = -L[-1] / L_target[n]
    objective += cost_term + 20* penalty_term

# Solve the problem
prob = cp.Problem(cp.Minimize(objective), constraints)
prob.solve(solver="CLARABEL")

# Collect data for plotting
d_history = np.array([x[n][:H].value for n in range(N)])
f_history = np.array([x[n][H:2*H].value for n in range(N)])
L_history = np.array([x[n][2*H:].value for n in range(N)])

# Plotting
fig, axes = plt.subplots(N, 1, figsize=(10, 6), sharex=True)

for n in range(N):
    axes[n].plot(d_history[n], label='d (drawn)', marker='o')
    axes[n].plot(f_history[n], label='f (fed)', marker='x')
    axes[n].plot(L_history[n], label='L (load)', marker='s')
    axes[n].set_title(f'Agent {n} Energy History')
    axes[n].legend()
    axes[n].grid(True)

axes[-1].set_xlabel('Time Horizon (h)')
plt.tight_layout()
plt.show()
```
